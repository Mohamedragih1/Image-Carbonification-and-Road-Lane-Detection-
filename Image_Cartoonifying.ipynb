```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/main image.png')
cv2_imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
denoised_image = cv2.medianBlur(gray_image, 9)
cv2_imshow(denoised_image)

laplacian = cv2.Laplacian(denoised_image, cv2.CV_64F, ksize=3)
cv2_imshow(laplacian)
```

```python
laplacian = cv2.Laplacian(denoised_image, cv2.CV_8U, ksize=5)
cv2_imshow(laplacian)
# Convert the edges to an 8-bit format and apply a binary threshold for a sketch effect
laplacian_edges_abs = cv2.convertScaleAbs(laplacian)
_, sketch = cv2.threshold(laplacian_edges_abs, 130, 255, cv2.THRESH_BINARY_INV)

cv2_imshow(sketch)
```

```python
#small_image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)
for i in range(5):
  image = cv2.bilateralFilter(image, d=9, sigmaColor=15, sigmaSpace=250)

# filtered_image = cv2.resize(filtered_small, (image.shape[1], image.shape[0]))
cv2_imshow(image)
```

```python
cartoon = cv2.bitwise_and(image, image, mask=sketch)
cv2_imshow(cartoon)
```
